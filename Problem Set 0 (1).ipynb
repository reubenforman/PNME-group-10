## Exercise 2

```python
str_x = "Facts about the distribution of wealth have been highlighted in a large number of studies, including \
Wolff (1992, 1998), Cagetti and De Nardi (2008), and Moritz and Rios-Rull (2015). A striking aspect of \
the wealth distribution in the US is its degree of concentration. Over the past 30 years or so, for instance, \
households in the top % of the wealth distribution have held about one-third of the total wealth in the economy, and \
those in the top 5% have held more than half. At the other extreme, more than 10% of households \
have little or no assets. While there is agreement that the share held by the richest few is very high, the \
extent to which the shares of the richest have changed over time (and why) is still the subject of some debate \
(Piketty 2014, Kopczuk 2014, Saez and Zucman 2014, and Bricker et al. 2015)."
```

```python
def count_str(str, data):
    count = data.lower().count(str.lower())
    return count

print(f"The number of times wealth appears in the text is {count_str('wealth', str_x)} times, the number of times distribution \
appears in the text is {count_str('distribution', str_x)} times, the number of times households appears in the text is \
{count_str('households', str_x)} times, the number of times assets appears in the text is {count_str('assets', str_x)} times.")
```

## Exercise 3

```python
import numpy as np
x = np.linspace(0, 10, 100)
y = np.log(x) # Encounters an error of log(0)
z = np.sin(x)
print(f"The type of y is {type(y)}, the size of y is {len(y)} and the size of z is {len(z)} which logically makes sense as\
the size of x is 100 and y and z are using the same number of inputs as x so should give the same output.")
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2, figsize=(15,5))

# Plot in the x-y plane
axes[0].plot(y, x)
axes[0].set_title("y-x")
axes[0].set_xlabel("y")
axes[0].set_ylabel("x")

# Plot in the x-z plane
axes[1].plot(z, x)
axes[1].set_title("z-x")
axes[1].set_xlabel("z")
axes[1].set_ylabel("x")

plt.tight_layout()
plt.show(close=None, block=None)
```

```python
import scipy as sp
from scipy.optimize import minimize_scalar
# Parameter values
alpha = 0.5
y = 10
p1 = 1
p2 = 2

def utility_function(x1, alpha, y, p1, p2):
    x2 = (y - p1*x1) / p2 # Income not spent on first good is spent on the second good
    utility = x1**alpha * x2**(1 - alpha)
    return utility

obj_func = lambda x1: -utility_function(x1, alpha, y, p1, p2)
res = minimize_scalar(obj_func)
print(f"The optimal x1 is {res.x} and the optimal x2 is {(y - p1*res.x) / p2}. Meaining the maximum utility is\
 {utility_function(res.x, alpha, y, p1, p2)}.")
```
